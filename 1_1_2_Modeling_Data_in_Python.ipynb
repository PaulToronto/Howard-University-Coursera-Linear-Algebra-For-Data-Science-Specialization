### Basic Example

```python
x = np.linspace(-3, 3, 21)
y = x**2
x, y
```

```python
len(x)
```

```python
plt.plot(x, y, 'g-*')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Basic Plot');
```

### Multiple Plots on Same Canvas

```python
# (number_rows, number_columns, index)
# the index starts at 1

plt.subplot(2, 2, 1)
plt.plot(x, x**2, 'g')
plt.title('$f(x) = x^{2}$')

plt.subplot(2, 2, 2)
plt.plot(x, x**3, 'r--')
plt.title('$f(x) = x^{3}$');

plt.subplot(2, 2, 3)
plt.plot(x, x**2, 'b')
plt.title('$f(x) = x^{2}$')

plt.subplot(2, 2, 4)
plt.plot(x, x**3, 'm-*')
plt.title('$f(x) = x^{3}$');

plt.tight_layout(); # ensures everything fits
```

### Introduction to the Object Oriented Method

- The dimensions [left, bottom, width, height] of the new axes. All quantities are in fractions of figure width and height.

```python
fig = plt.figure(figsize=(5, 4))
# rectangle [left, bottom, width, height]
# this makes more sense after looking at next example
axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x, y, 'b*');
```

```python
fig = plt.figure(figsize=(5, 4))

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'b-o', lw=2)
axes1.set_title('Larger Graph')

axes2 = fig.add_axes([0.2, 0.1, 0.2, 0.3])
axes2.plot(y, x, 'r->', lw=2)
axes2.set_title('Smaller Graph');
```

```python
fig = plt.figure(figsize=(5, 4))

# the smaller graph is hidden by the larger graph
axes2 = fig.add_axes([0.2, 0.1, 0.2, 0.3])
axes2.plot(y, x, 'r->', lw=2)
axes2.set_title('Smaller Graph')

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'b-o', lw=2)
axes1.set_title('Larger Graph');
```

```python
fig = plt.figure(figsize=(5, 4))
# zorder is set so the smaller graph is visible
axes2 = fig.add_axes([0.2, 0.1, 0.2, 0.3])
axes2.plot(y, x, 'r->', lw=2)
axes2.set_title('Smaller Graph')
axes2.zorder = 2

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'b-o', lw=2)
axes1.set_title('Larger Graph')
axes1.zorder = 1;
```

```python
fig = plt.figure(figsize=(5, 4))

axes1 = fig.add_axes([0.1, 0.1, 0.4, 0.8])
axes1.plot(x, y, 'b-o', lw=2)
axes1.set_title('Left Graph')

axes2 = fig.add_axes([0.6, 0.1, 0.3, 0.8])
axes2.plot(x, y, 'r->', lw=4)
axes2.set_title('Right Graph');
```

### `subplots()`

- the `plt.subplots()` object will act as a more automatic axis manager

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# Empty canvas of 1 by 3 subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))

for index, ax in enumerate(axes):
    ax.plot(x, x**(index), 'm')
    ax.set_xlabel('x-label')
    ax.set_ylabel('y-label')
    ax.set_title(f'$f(x) = x^{index}$')

axes[1].set_title("I have overidden the title of this ax!");
```

#### Legends

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, 'r*', label='$f(x) = x^{2}$')
ax.plot(x, x**3, 'g--', label='$f(x) = x^{3}$')
ax.legend(loc=4);

# loc=0, best
# loc=1, top right
# loc=2, top left
# loc=3, bottom left
# loc=4, bottom right
```
