### Basic Example

```python
x = np.linspace(-3, 3, 21)
y = x**2
x
```

```python
len(x)
```

```python
plt.plot(x, y, 'g-*')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('Basic Plot');
```

### Multiple Plots on Same Canvas

```python
# (number_rows, number_columns, index)

plt.subplot(1, 2, 1)
plt.plot(x, y, 'g')
plt.title('$f(x) = x^{2}$')

plt.subplot(1, 2, 2)
plt.plot(x, x**3, 'r--')
plt.title('$f(x) = x^{3}$');
```

### Introduction to the Object Oriented Method

- The dimensions [left, bottom, width, height] of the new axes. All quantities are in fractions of figure width and height.

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x, y, 'b*');
```

```python
fig = plt.figure(figsize=(10, 8))

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'b-o', lw=2)
axes1.set_title('Larger Graph')

axes2 = fig.add_axes([0.3, 0.1, 0.2, 0.3])
axes2.plot(x, y, 'r->', lw=4)
axes2.set_title('Smaller Graph');
```

```python
fig = plt.figure(figsize=(10, 8))

axes1 = fig.add_axes([0.1, 0.1, 0.4, 0.8])
axes1.plot(x, y, 'b-o', lw=2)
axes1.set_title('Left Graph')

axes2 = fig.add_axes([0.6, 0.1, 0.3, 0.8])
axes2.plot(x, y, 'r->', lw=4)
axes2.set_title('Right Graph');
```

### `subplots()`

- the `plt.subplots()` object will act as a more automatic axis manager

```python
fig, axes = plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

```python
# Empty canvas of 1 by 3 subplots
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(12, 4))
```

```python
# axes is an array of axes to plot on
axes
```

```python
for index, ax in enumerate(axes):
    print(index, ax)
```

```python
for index, ax in enumerate(axes):
    ax.plot(x, (index + 1) * x**2, 'm')
    ax.set_xlabel('x-label')
    ax.set_ylabel('y-label')
    ax.set_title('$f(x) = ' + str(index + 1) + 'x^2$')

fig  
```

### Legends, labels and titles

#### Figure titles

- a title can be added to each axis instance in a figure
- use the `set_title` method of the axes instance

```python
for index, ax in enumerate(axes):
    ax.plot(x, (index + 1) * x**2, 'm')
    ax.set_xlabel('x-label')
    ax.set_ylabel('y-label')
    
    # ax.set_title
    ax.set_title('$f(x) = ' + str(index + 1) + 'x^2$')

fig  
```

#### Axis labels

- set the labels of the x and y axis

```python
for index, ax in enumerate(axes):
    ax.plot(x, (index + 1) * x**2, 'm')

    # Axis labels
    ax.set_xlabel('x' + str(index))
    ax.set_ylabel('y' + str(index))
    
    ax.set_title('$f(x) = ' + str(index + 1) + 'x^2$')

fig  
```

#### Legends

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, 'r*', label='$f(x) = x^{2}$')
ax.plot(x, x**3, 'g--', label='$f(x) = x^{3}$')
ax.legend(loc=4);

# loc=0, best
# loc=1, top right
# loc=2, top left
# loc=3, bottom left
# loc=4, bottom right
```
