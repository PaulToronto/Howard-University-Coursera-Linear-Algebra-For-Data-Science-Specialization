# 3.3 Understanding Linear Regression Models

## Imports

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

## Multiple Linear Regression in Python

- be careful that the variables are not higly correlated with each other
    - maybe drop one or more features
- each feature should be normally distributed with a mean near 0

### The Data

```python
path = 'https://raw.githubusercontent.com/PaulToronto/'
path += 'Howard-University-Coursera-Linear-Algebra-For-Data-Science-Specialization/'
path += 'main/data/real_estate_price_size_year_view.csv'

data = pd.read_csv(path)
data.head()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data = pd.get_dummies(data, 'view', drop_first=True)
data
```

```python
x1 = data[['size', 'year', 'view_Sea view']]
y = data['price']
```

```python
x1
```

```python
y
```

### Regression

```python
x = sm.add_constant(x1)
x
```

```python
results = sm.OLS(y, x).fit()
results.summary()
```
